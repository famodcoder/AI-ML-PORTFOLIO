```python
def calculate_factorial(n):
    # Factorials aren't defined for negative numbers
    if n < 0:
        return "Error: Please enter a non-negative integer."

    result = 1

    # We use a 'for' loop to multiply from 1 up to n
    # range(1, n + 1) gives us numbers starting at 1 and ending at n
    for i in range(1, n + 1):
        result = result * i

    return result

# Get user input
num = int(input("Enter a number to find its factorial: "))
print(f"The factorial of {num} is {calculate_factorial(num)}")
```
